### Write two Python programs to demonstrate that you are able to read some image files from your Google Drive and then process those images.

```python
# Import image files

import os
import matplotlib.pyplot as plt
import numpy as np
import scipy
import cv2

# Mount Google Drive to Google Colab
from google.colab import drive
drive.mount('/content/drive')

# Update the working directory to access image
os.chdir('/content/drive/My Drive/Colab Notebooks/Pictures')

# Import images into CoLab as grayscale 2D arrays
img1 = cv2.imread('Ballin_Cat.png', 0)
img2 = cv2.imread('Pepper.png', 0)
```

```python
# Median Filter

# Convert image to a numpy array for processing
img1_np = np.array(img1)
# Apply median filter: The size of the filter is 50x50. This means that
# a median value will be calculated for a 50x50 area around each element.
img1_median = scipy.ndimage.median_filter(img1_np, size = (25, 25))

# Display original image
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(img1_np, cmap = 'gray')
# Display filtered image
plt.subplot(1, 2, 2)
plt.title('Median Filtered Image')
plt.imshow(img1_median, cmap = 'gray')
```

```python
# Low Pass Gaussian Filter

# Convert image to a numpy array for processing
img2_np = np.array(img2)

img2_lp = scipy.ndimage.gaussian_filter(img2_np, sigma = 15)

# Display original image
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(img2_np, cmap = 'gray')
# Display filtered image
plt.subplot(1, 2, 2)
plt.title('Gaussian Filtered Image')
plt.imshow(img2_lp, cmap = 'gray')
```
